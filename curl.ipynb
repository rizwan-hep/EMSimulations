```python
from sympy.vector import CoordSys3D, vector_integrate, curl
from sympy.geometry import Circle, Point
from sympy import pi
from spb import plot_vector
R=CoordSys3D('R')


A_x=-R.y**3
A_y=R.x**3
A = A_x * R.i + A_y * R.j

print("The vector field we are dealing with is:", A)
plot_vector(A, (R.x, -2, 2), (R.y, -2, 2), adaptive=False,
quiver_kw=dict(color="black", scale=150, headwidth=4),
contour_kw={"cmap": "rainbow", "levels": 100},
grid=True, xlabel="x", ylabel="y", title="$A = -y^3 \\hat{i} + x^3 \\hat{j}$")

epsilon=1.0e-6 #Just an arbitrary but small number. This is for convergence

radius=10 #Reset radius to 2
print("We begin by centering the circle at the point (0,0)")
print("We are using a starting radius of 10")
while radius>epsilon:
    print("The radius is:", radius)
    circle = Circle(Point(0,0), radius)
    print("The estimate of the magnitude of curl is:", (vector_integrate(A,  circle))/(pi*radius**2))
    radius=radius/10
print("It seems obvious that the curl is converging to zero at the origin")
curl_A = curl(A)
print("The expression for the curl of this field is: ", curl_A)
print("Evaluating the curl at the origin gives:", curl_A.subs({R.x:0, R.y:0}))
print("")
print("")
print("We now center the circle at the point (0,1)")

radius=10 #Reset radius to 2
while radius>epsilon:
    circle = Circle(Point(0,1), radius)
    print("The radius is:", radius)
    print("The estimate of the magnitude of curl is:", (vector_integrate(A,  circle))/(pi*radius**2))
    radius=radius/10
print("It seems obvious that the magnitude of curl is converging to 3.0 at the point(0,1)")
print("Evaluating the curl at the point (0,1) gives:", curl_A.subs({R.x:0, R.y:1}))

```

```python
A_x=-R.y**2
A_y=R.x**2
A = A_x * R.i + A_y * R.j

print("The vector field we are dealing with is:", A)
plot_vector(A, (R.x, -2, 2), (R.y, -2, 2), adaptive=False,
quiver_kw=dict(color="black", scale=150, headwidth=4),
contour_kw={"cmap": "rainbow", "levels": 100},
grid=True, xlabel="x", ylabel="y", title="$A = -y^2 \\hat{i} + x^2 \\hat{j}$")


epsilon=1.0e-6 #Just an arbitrary but small number. This is for convergence


radius=2 #Reset radius to 2
print("We begin by centering the circle at the point (0,0)")
print("We are using a starting radius of 2")
circle = Circle(Point(0,0), radius)
print("The estimate of the magnitude of curl is:", (vector_integrate(A,  circle))/(pi*radius**2))
print("Visual inspection of the vector field should convince us of this result")
print("")
print("")
print("We now center the circle at the point (0,1)")

radius=10 #Reset radius to 2
while radius>epsilon:
    
    circle = Circle(Point(0,1), radius)
    print("The radius is:", radius)
    print("The estimate of the magnitude of curl is:", (vector_integrate(A,  circle))/(pi*radius**2))
    radius=radius/10

print("It seems obvious that the magnitude of curl is converging to 2.0 at the point(0,1)")
curl_A = curl(A)
print("The expression for the curl of this field is: ", curl_A)
print("Evaluating the curl at the point (0,1) gives:", curl_A.subs({R.x:0, R.y:1}))
print("This result requires further elaboration. Notice that the curl estimate did not")
print("change as we reduced the radius of the circle. This is because of the special nature")
print("of the vector field we are dealing with. This is certainly not true in general,")
print(" as we have already seen. ")

```

```python
from sympy.abc import phi
from sympy import pi, cos, sin
from sympy.vector import ParametricRegion

A_x=-R.y**4
A_y=2*R.z**4
A_z=3*R.x**5  
A = A_x * R.i + A_y * R.j + A_z * R.k

curl_A = curl(A)
print("The expression for the curl of this field is: ", curl_A)
print("Evaluating the curl at the point (1,1,1) gives:", curl_A.subs({R.x:1, R.y:1, R.z:1}))
print("We will now attempt to estimate this numerically.")
print("We need three loops to find the three components of the curl, ")
print("a loop perpendicular to each direction in which we wish to find the curl")
print("")
x_0=1 #The x, y, and z coordinates of the center of the circle
y_0=1
z_0=1
print("We begin with the z component")
radius=epsilon #Reset radius to 2
circle = ParametricRegion((x_0+radius*cos(phi), y_0+radius*sin(phi), z_0), (phi, 0, 2*pi))
print("The estimate of the magnitude of curl in the z-direction is:", (vector_integrate(A,  circle))/(pi*radius**2))

print("Now the x component")
circle = ParametricRegion((x_0, y_0+radius*cos(phi), z_0+radius*sin(phi)), (phi, 0, 2*pi))
print("The estimate of the magnitude of curl in the x-direction is:", (vector_integrate(A,  circle))/(pi*radius**2))

print("Now the y component")
circle = ParametricRegion((x_0+radius*sin(phi), y_0, z_0+radius*cos(phi)), (phi, 0, 2*pi))
print("The estimate of the magnitude of curl in the y-direction is:", (vector_integrate(A,  circle))/(pi*radius**2))

print("")
x_0=1
y_0=2
z_0=3
print("Let's attempt to find the curl at the point (1,2,3) whose exact value is:", curl_A.subs({R.x:x_0, R.y:y_0, R.z:z_0}))
circle = ParametricRegion((x_0+radius*cos(phi), y_0+radius*sin(phi), z_0), (phi, 0, 2*pi))
print("The estimate of the magnitude of curl in the z-direction is:", (vector_integrate(A,  circle))/(pi*radius**2))

print("Now the x component")
circle = ParametricRegion((x_0, y_0+radius*cos(phi), z_0+radius*sin(phi)), (phi, 0, 2*pi))
print("The estimate of the magnitude of curl in the x-direction is:", (vector_integrate(A,  circle))/(pi*radius**2))

print("Now the y component")
circle = ParametricRegion((x_0+radius*sin(phi), y_0, z_0+radius*cos(phi)), (phi, 0, 2*pi))
print("The estimate of the magnitude of curl in the y-direction is:", (vector_integrate(A,  circle))/(pi*radius**2))
```
